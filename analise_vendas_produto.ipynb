```python
# ANÁLISE DE VENDAS POR PRODUTO

# Importa a biblioteca 
import pandas as pd
from IPython.display import display


# Carrega o dataset limpo a partir do arquivo CSV
df = pd.read_csv("data_clean.csv")

# Exibe as primeiras linhas para conferência
display(df.head())


# CRIA A COLUNA TOTAL_VENDA

# Calcula o total da venda multiplicando
# a quantidade pelo preço unitário
df["Total_Venda"] = (df["Quantidade"] * df["Preço"]).round(2)

# Exibe o dataframe com a nova coluna
print("\nDataFrame com a coluna Total_Venda:")
display(df.head())


# CALCULA O TOTAL DE VENDAS POR PRODUTO

# Agrupa os dados por Produto
# e soma o valor total das vendas
vendas_por_produto = (
    df.groupby("Produto")["Total_Venda"]
    .sum()
    .round(2)
    .reset_index()
)

# Exibe o total de vendas por produto
print("\nTotal de vendas por produto:")
display(vendas_por_produto)


# IDENTIFICA O PRODUTO COM MAIOR VALOR DE VENDAS TOTAIS

# Ordena os produtos pelo total de vendas
# em ordem decrescente
vendas_ordenadas = vendas_por_produto.sort_values(
    by="Total_Venda",
    ascending=False
)

# Seleciona o produto com maior valor de vendas
produto_mais_vendido = vendas_ordenadas.iloc[0]

# Exibe o resultado
print("\nProduto com maior valor total de vendas:")
display(produto_mais_vendido.to_frame().T)
```
